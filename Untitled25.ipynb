```python

import pandas as pd
import matplotlib.pyplot as plt


df = pd.read_csv('/content/Churn_Modelling.csv')


gender_counts = df['Gender'].value_counts()


plt.figure(figsize=(4, 4))
plt.pie(gender_counts, labels=gender_counts.index, autopct='%1.1f%%', startangle=90, colors=['#ff9999','#66b3ff'])
plt.title('Customer Distribution by Gender')
plt.axis('equal')
plt.show()
```

```python

!pip install matplotlib pandas

import pandas as pd
import matplotlib.pyplot as plt


df = pd.read_csv('/content/Churn_Modelling.csv')


age_counts = df['Age'].value_counts().sort_index()
churn_rate_by_age = df.groupby('Age')['Exited'].mean().sort_index()

# Plotting the line plots
fig, ax1 = plt.subplots(figsize=(10, 5))

ax1.set_xlabel('Age')
ax1.set_ylabel('Number of Customers', color='tab:blue')
ax1.plot(age_counts.index, age_counts.values, color='tab:blue', marker='o', label='Number of Customers')
ax1.tick_params(axis='y', labelcolor='tab:blue')


ax2 = ax1.twinx()
ax2.set_ylabel('Churn Rate', color='tab:red')
ax2.plot(churn_rate_by_age.index, churn_rate_by_age.values, color='tab:red', marker='x', linestyle='--', label='Churn Rate')
ax2.tick_params(axis='y', labelcolor='tab:red')


fig.suptitle('Customers and Churn Rate by Age')
ax1.legend(loc='upper left')
ax2.legend(loc='upper right')


plt.show()
```

```python
import pandas as pd
from wordcloud import WordCloud
import matplotlib.pyplot as plt

df = pd.read_csv('/content/Churn_Modelling.csv')

text = df['Surname'] + ' ' + df['Geography'] + ' ' + df['Gender']

wordcloud = WordCloud(width=800, height=400, background_color='white').generate(' '.join(text))

# Plot the word cloud
plt.figure(figsize=(10, 5))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('/content/Churn_Modelling.csv')


grouped_data = df.groupby(['Geography', 'Gender'])['Exited'].mean().reset_index()

# Plotting the bar graph
plt.figure(figsize=(10, 6))
plt.bar(grouped_data['Geography'] + ' - ' + grouped_data['Gender'], grouped_data['Exited'] * 100, color=['blue', 'green', 'red', 'purple'])
plt.xlabel('Geography - Gender')
plt.ylabel('Churn Rate (%)')
plt.title('Churn Rate by Geography and Gender')
plt.xticks(rotation=45, ha='right')
plt.ylim(0, 100)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt


df = pd.read_csv('/content/Churn_Modelling.csv')

grouped_data = df.groupby(pd.cut(df['EstimatedSalary'], bins=10))['Exited'].mean().reset_index()

# Plotting the bar graph
plt.figure(figsize=(10, 6))
plt.bar(grouped_data['EstimatedSalary'].astype(str), grouped_data['Exited'] * 100, color='skyblue')
plt.xlabel('Estimated Salary (Grouped)')
plt.ylabel('Churn Rate (%)')
plt.title('Churn Rate by Estimated Salary')
plt.xticks(rotation=45, ha='right')
plt.ylim(0, 100)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns


df = pd.read_csv('/content/Churn_Modelling.csv')


churn_by_geography = df.groupby('Geography')['Exited'].mean().reset_index()

# Plotting the heatmap
plt.figure(figsize=(10, 6))
heatmap_data = pd.pivot_table(churn_by_geography, values='Exited', index='Geography', columns='Geography')
sns.heatmap(heatmap_data, annot=True, cmap='YlOrRd', fmt='.2%', linewidths=.5)
plt.xlabel('Churn Rate')
plt.ylabel('Geography')
plt.title('Churn Rate by Geography')
plt.yticks(rotation=0)
plt.tight_layout()
plt.show()
```
